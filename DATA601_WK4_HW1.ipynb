## and *Parse* XML

As a string input, we'll use xml elementree to parse the html

```
#Use xml etree
import xml.etree.ElementTree as ET
tree = ET.ElementTree(ET.fromstring(str_text))
root = tree.getroot()
```

## Find and return all of the HTML data (children.text)

```
#The key here was children.text, which prints the HTML
i=0

for child in root:
    while i < 2:
        for children in child:
            print(children.text)
            i+=1
```

## Step 2: Use BeautifulSoup to Parse

## Beginning to parse with beautifulsoup
This looks good. We have a mostly-clean output of the webpages, although there are some elements that are not necessary (dates, etc.)

Let's move forwards, and use beautiful soup here to create a list of websites and elements.

```
#The key here was children.text, which prints the HTML
full_text = []

for child in root:
    for children in child:
        full_text.append(children.text)
```

```
#How many individual elements are there?
len(full_text)
```

This doesn't necessarily mean the websites. There are definitely elements in here that are not websites.

```
#Parse each individual page from the list
all_links = []

for page in full_text:
    page_soup = BeautifulSoup(page)
    
    page_links = []
    for a in page_soup.find_all('a', href=True):
        page_links.append(a['href'])

       #print("Found the URL Link:", a['href'])
    all_links.append(page_links)

all_links
```

## Let's clean up the list:
- Some of the elements that were added to the list were not part of a full website, so we should remove them (there are elements such as dates, etc.)

- Remove all empty elements
https://www.geeksforgeeks.org/python-remove-empty-list-from-list/
- Uses list comprehension.
- Checks to see if any lists are empty, and only keeps the lists with elements as list 'cleaned_links'

```
#Cleaned Links
cleaned_links = [x for x in all_links if x != []]
```

### Finally, let's create an output of each website's links:
*Website n has x links: (list)*

```
#Shows each website, and how many links it has
counter = 0

for website in cleaned_links:
    counter += 1
    print(f"Website {counter} has {len(website)} links:")
    
    for links in website:
        print(links)
    
    print("\n")
```
