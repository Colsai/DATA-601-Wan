This is a short parser for each of the cells.
- This function finds the code cells, 
- finds the source of the code cells, 
- and adds functions based on the required 'import x', 
- If not comment line (startswith #)
- Split line into elements, access second element

(For more explanation, see end)

```python
def parse_packages(nbk):

    #Find and convert to code Cells
    code_cells = [cell for cell in nbk['cells'] if cell['cell_type'] == 'code'] #Look for code cells
    code_lines = [elem['source'] for elem in code_cells] #Find the 'source' == raw code cells

    imports_list = [] #Create list to keep imported packages

    for cell in code_lines:
        for line in cell:
            if ("import" in line.lower()) and (line.startswith("#") == False):
                line_items = line.split()
                imports_list.append(line_items[1])

    return imports_list
```

This is a function for lines (I separated it out as a different function).
- It looks at only code cells

```python
def count_code_lines(nbk):
    code_cells = [cell for cell in nbk['cells'] if cell['cell_type'] == 'code'] #Look for code cells
    code_lines = [elem['source'] for elem in code_cells] #Find the 'source' == raw code cells

    line_counter = 0

    for line in code_lines:
        for elem in line:
            if not elem.strip():
                continue

            else:
                line_counter+=1
                #print(f"{line_counter}. {elem}")
    
    return int(line_counter)
```

And running it through the notebook...

```python
all_pckgs = []

for notebook in py_nbks:
    f = open(notebook, "r")
    nbk_json = json.loads(f.read())

    packages_used = parse_packages(nbk_json) #Returns a list of packages used as function
    [all_pckgs.append(pckg) for pckg in packages_used if pckg not in all_pckgs] #As list comp

all_pckgs
```
